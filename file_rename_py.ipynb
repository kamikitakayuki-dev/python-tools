```python
import os

folder = "test"
os.makedirs(folder, exist_ok=True)

for i, filename in enumerate(os.listdir(folder)):
    old_path = os.path.join(folder, filename)
    new_path = os.path.join(folder, f"file_{i}.txt")
    os.rename(old_path, new_path)

print("リネーム完了")
```
